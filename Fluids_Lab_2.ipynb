```
import plotly.graph_objects as go
import numpy as np
import math as m
import matplotlib.pyplot as plt
```

```
# Table 1 slopes of vanes

_120degCone = 1.5
flatPlate = 1 
hemisphere = 2 
_30degCone = 0.87

```

```
# Table 2

A = 78.54 #mm2 nominal nozzle area
M = 600 #g mass of jockey
s = 35 #mm distance from nozzle tip to the impact point on the vane

rho = 1000 #kg/m3
g = 9.81 #m/s2


#Adjust units for SI
A = 78.54 / 1e6#[m2]
M = 600 / 1e3 # [kg]
s = 35 / 1000 #[m]
```

```
# Table 3

Q = np.array([11, 15.1, 20, 24.8, 29, 31.2,])/60/1000 #m3/s data collected was in Liters per minute => convert to m3/s
y = np.array([7, 19, 31, 51, 71, 82])/1000 # [m]
B = m.radians(90)

mfr = rho * Q 
# nozzle exit velocity, vane contact velocity, mass flow rate, and mass flow rate times vane contact velocity
u = Q / A
u0 = np.sqrt(u**2 - 2*g*s)
mfru0 = mfr * u0

#Experimental and Theoretical force on vane
eF = M*g*y/0.15
tF = mfru0*(1 - m.cos(B))

tF = rho * Q * np.sqrt((Q / A)**2 - 2*g*s) 
# print(rho,A,s,g,'\n',Q )

mu1 = mfru0
eF1 = eF
tF1 = tF

fig = go.Figure(data=[go.Table(header=dict(values=['$Q\\, [m^{3}/s]$', '$y \\,(m)$','$\dot{m}\\,[kg/s]$',
                                                   '$u\\, [m/s]$','$u_{0}\\,[m/s]$','$\dot{m}u_{0}\\,[kg\cdot m\cdot s^{-2}]$',
                                                   '$Experimental\\,Force\\,,\\,F\\,(N)$',
                                                   '$Theoretical\\,Force\\,,\\,F\\,(N)$']),
                               cells=dict(values=[np.around(Q,6),y,np.around(mfr,6),np.around(u,4),np.around(u0,4),
                                                  np.around(mfru0,4), np.around(eF,4),np.around(tF,4)]))])
fig.update_layout(title_text='Table 3')
fig.show()
```

```
# Table 4

Q = np.array([10.5, 15.1, 20, 24.5, 30, 32.2])/1000/60 #m3/s
y = np.array([10, 25, 48, 75, 120, 136])/1000 #m
B = m.radians(120)

mfr = rho * Q 
# nozzle exit velocity, vane contact velocity, mass flow rate, and mass flow rate times vane contact velocity
u = Q / A
u0 = np.sqrt(u**2 - 2*g*s)
mfru0 = mfr * u0

#Experimental and Theoretical force on vane
eF = M*g*y/0.15
tF = mfru0*(1 - m.cos(B))
# print(rho,A,s,g,'\n',Q )

mu2 = mfru0
eF2 = eF
tF2 = tF

fig = go.Figure(data=[go.Table(header=dict(values=['$Q\\, [m^{3}/s]$', '$y \\,(m)$','$\dot{m}\\,[kg/s]$',
                                                   '$u\\, [m/s]$','$u_{0}\\,[m/s]$',
                                                   '$\dot{m}u_{0}\\,[kg\cdot m\cdot s^{-2}]$',
                                                   '$Experimental\\,Force\\,,\\,F\\,(N)$',
                                                   '$Theoretical\\,Force\\,,\\,F\\,(N)$']),
                               cells=dict(values=[np.around(Q,6),y,np.around(mfr,6),
                                                  np.around(u,4),np.around(u0,4),np.around(mfru0,4),
                                                  np.around(eF,4),np.around(tF,4)]))])
fig.update_layout(title_text='Table 4')
fig.show()
```

```
# Table 5 

Q = np.array([10, 15, 20, 25, 30, 32.2])/1000/60 #m3/s
y = np.array([12, 32, 60, 100, 143, 165])/1000 #m
B = m.radians(180)

mfr = rho * Q 
# nozzle exit velocity, vane contact velocity, mass flow rate, and mass flow rate times vane contact velocity
u = Q / A
u0 = np.sqrt(u**2 - 2*g*s)
mfru0 = mfr * u0

#Experimental and Theoretical force on vane
eF = M*g*y/0.15
tF = mfru0*(1 - m.cos(B))
# print(rho,A,s,g,'\n',Q )

mu3 = mfru0
eF3 = eF
tF3 = tF

fig = go.Figure(data=[go.Table(header=dict(values=['$Q\\, [m^{3}/s]$', '$y \\,(m)$',
                                                   '$\dot{m}\\,[kg/s]$','$u\\, [m/s]$',
                                                   '$u_{0}\\,[m/s]$','$\dot{m}u_{0}\\,[kg\cdot m\cdot s^{-2}]$',
                                                   '$Experimental\\,Force\\,,\\,F\\,(N)$',
                                                   '$Theoretical\\,Force\\,,\\,F\\,(N)$']),
                               cells=dict(values=[np.around(Q,6),y,np.around(mfr,6),np.around(u,4),
                                                  np.around(u0,4),np.around(mfru0,4), np.around(eF,4),
                                                  np.around(tF,4)]))])
fig.update_layout(title_text='Table 5')
fig.show()
```

```
# Plot for 90 degree vane


b1, b0 = np.polyfit(mu1, eF1, 1)
fit = b0 + b1*mu1
plt.plot(mu1,eF1,'ro', label='Experimental Force')
plt.plot(mu1,tF1,'bo', label='Theoretical Force')
plt.plot(mu1, fit)
plt.xlabel(' rate of delivery, $\dot{m}u_{0}\\,[kg\cdot m\cdot s^{-2}]$')
plt.ylabel('Force, F [N]')
plt.title('Rate of Delivery vs Force,  $\\beta\\,=\\,90\\degree$')
plt.legend(loc='lower right')
plt.gcf().set_size_inches(10,7)
plt.show()

b90 = b1
```

```
eb = np.array([b90,b120,b180])
b90t,b0 = np.polyfit(mu1, tF1,1)
b120t,b0 = np.polyfit(mu2, tF2,1)
b180t,b0 = np.polyfit(mu3, tF3,1)
tb = np.array([b90t,b120t, b180t])

dev = ((tb - eb)/tb)*100 #Calculate Deviation

fig = go.Figure(data=[go.Table(header=dict(values=['$\\beta\\,(degrees)$', 
                                                   '$Gradient,\\,m_t\\,(linear fit)$',
                                                   '$Gradient,\\,m_f\\,(theory)$',
                                                   '$Deviation,\\,\\frac{m_{t}-m_{f}}{m_{t}}\\,(\\%)$']),
                               cells=dict(values=[['$90$','$120$','$180$'],
                                                  np.around(eb,6),np.around(tb,6),
                                                  np.around(dev,4)]))])
fig.update_layout(title_text='Table 6')
fig.show()
```
