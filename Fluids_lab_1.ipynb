```
import matplotlib.pyplot as plt
import numpy as np 
import math as m
  
```

```
#Table 1

l = np.array([-54,-34, -22, -8, 7, 22, 37, 52, 67, 82, 102]) #[mm]
d = np.array([26, 23.2, 18.4, 16, 16.79, 18.47, 20.16, 21.84, 23.53, 25.21, 26]) #[mm]
A = np.array([530.9, 422.7, 265.9, 201.1, 221.4, 267.9, 319.2, 374.6, 434.8, \
              499.2, 530.9])/1e6 #[m2]
  
d2 = d[3]
d1 = d[0]

#d2/dn ratio
# print('d2/dn ratio =', d[3]/d)

#from equation 4 on the lab handout
#ideal dimensionless pressure
idp = (d2/d1)**4 - (d2/d)**4
# print('\nideal dimensionless pressure =', idp)


import plotly.graph_objects as go

fig = go.Figure(data=[go.Table(header=dict(values=['$Tapping$', '$l\\,(mm)$','$d\\,(mm)$','$A\\,(m^2)$','$\\large\
  \\frac{d_2}{d_n}$','$Ideal\\,Dimensionless\\,Pressure$']),cells=dict(values=[['$A\\,(upstream, h_1)$','$B$','$C\
  $','$D\\,(throat, h_2)$','$E$','$F$','$G$','$H$','$J$','$K$','$L$'], l,d,numpy.around(A,6),numpy.around(d[3]/d,6),numpy.around(idp,6)]))])
fig.update_layout(title_text='Table 1')
fig.show()

```

```
#Table 2 

Q1 = 5 / 19.38 / 1000 #[m3/s]
u1 = Q1 / A[0] #[m/s]
g = 9.81 #[m/s2]

#data collected
h = np.array([202, 192, 158, 42, 64, 112, 140, 157, 166, 172, 176])/1000 #[m]

#from equation 2 on the lab handout
# u2 = m.sqrt(2 * g * (u1**2 / (2 * g) + h[0] - h[3] )) #[m/s]
 
u2 = Q1 / A[3]
  
print('Q1 =',Q1,'m3/s')
print('u2 =', u2, 'm/s')
print('u2^2/2g =',u2**2/(2*g),'m')

# hn - h1
# print('hn - h1 =', h - h[0]) #[m]

#actual dimensionless pressure 
#using eq 4
adp1 = (h - h[0]) / (u2**2 / (2 * g))
# print('actual dimensionless pressure =', adp)

fig = go.Figure(data=[go.Table(header=dict(values=['$Tapping$', '$l\\,(mm)$','$h_n\\,(m)$','$h_n - h_1\\,\
  (m)$','$Actual\\,Dimensionless\\,Pressure$']),cells=dict(values=[['$A\\,(upstream, h_1)$','$B$','$C$',\
  '$D\\,(throat,h_2)$','$E$','$F$','$G$','$H$','$J$','$K$','$L$'], l,h,numpy.around(h-h[0],6),numpy.around(adp1,6)]))])
fig.update_layout(title_text='Table 2')
fig.show()
```

```
#Table 3 
h = np.array([298,294,262,210,220,248,262,272,276,282,284])/1000 #[m]
Q2 = 5 / 22.99 / 1000 #[m3/s]
print('Q2 = ',Q2,'m3/s')
u2 = Q2 / A[3] #[m/s]  
print('u2 =', u2,'m/s')
print('u2^2/2g =',u2**2/(2*g),'m')

#hn - h1
# print('hn - h1 = ', h - h[0])

#actual dimensionless pressure 
#using eq 4
adp2 = (h - h[0]) / (u2**2 / (2 * g))
# print('actual dimensionless pressure =', adp)

fig = go.Figure(data=[go.Table(header=dict(values=['$Tapping$', '$l\\,(mm)$','$h_n\\,(m)$','$h_n - h_1\\,(m)$','$A\
  ctual\\,Dimensionless\\,Pressure$']),cells=dict(values=[['$A\\,(upstream, h_1)$','$B$','$C$','$D\\,(throat, h_2)\
  $','$E$','$F$','$G$','$H$','$J$','$K$','$L$'], l,h,numpy.around(h-h[0],6),numpy.around(adp2,6)]))])
fig.update_layout(title_text='Table 3')
fig.show()
```

```
plt.plot(l, adp1, 'b.-')
plt.plot(l, adp2, 'c.-')
plt.plot(l,idp, 'r.-')
plt.ylabel('Dimensionless Pressure')
plt.xlabel('l (mm), Distance from Datum')
plt.legend(['Actual Dimensionless Pressre 1', 'Actual Dimensionless Pressre 2', 'Ideal Dimensionless Pressure'])
plt.title('Figure 1: Dimensionless Pressure vs Distance from Venturi Throat')
plt.gcf().set_size_inches(12, 6)
plt.show()
```
